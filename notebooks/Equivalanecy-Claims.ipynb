```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from carbonplan import styles

styles.mpl.set_theme(style='carbonplan_light', font_scale=1.5)
```

```python
method_display_names = {'mc': 'Moura Costa', 'ipcc': 'Lashof'}

data = pd.read_csv('/tmp/equivalency_data.csv')
data['method'] = data['method'].map(method_display_names)
```

```python
g = sns.FacetGrid(data=data, height=4, col='integration_time', hue='method', xlim=(0,100), ylim=(0.4, 1.05))
g.map(plt.plot, 'delay_length', 'ratio', lw=3.5)
axes = g.axes.flatten()
axes[0].set_ylabel('Relative Value of Delay')
#titles_d = {'mc': 'Moura Costa', 'ipcc': 'Lashoff'}
for ax in axes:
    title = ax.get_title()
    #ax.set_title(titles_d.get(title.split(' ')[-1], 'null'))
    ax.set_title(None)
    ax.set_xlabel('Delay Length')
    
plt.legend(bbox_to_anchor=(1,1), frameon=False, title='Method')
plt.grid(axis='y')
```
